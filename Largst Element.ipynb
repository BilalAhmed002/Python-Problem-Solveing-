```python
def largest_numberlist(inputlist: list[int]) -> int:
    """
    Function : This function is use for print largest Number

    arg: this funtion take only list 

    
    """
    largest = inputlist[0]
    for n in inputlist:
        if n > largest:
            largest = n            
    return largest
    
```

```python
def largest_numberlist(inputlist: list[float]) -> float:
    """
    Function : This function is use for print largest Number

    arg: this funtion take only list 

    
    """
    if not isinstance(inputlist,list):
        raise TypeError("Value should be Intiger")
        
    if len(inputlist)==0: 
        raise ValueError("List should not be Empty")   
        
    for i in inputlist:
        if not isinstance(i,(int,float)):
            raise TypeError("All elemnet must be int or flaot")
    

    
    largest = inputlist[0]
    for n in inputlist:
        if n > largest:
            largest = n            
    return largest
    
if __name__ =="__main__":

    try: 
        user_input=input("Enter the Value with space")
        inputlist = [float(item) for item in user_input.split()]
        
        result = largest_numberlist(inputlist)   
        print("largest : ",result)

    except (TypeError, ValueError) as e :
        print(e)
    except Exception as e :
        print(e)


    
```
