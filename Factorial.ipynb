```python
def findfactoral(num : int) -> int:
    """
    """
    if not isinstance(num,int):
        raise TypeError("Value must be intiger")
    if num <2 : 
        raise ValueError ("Input must be greater then 2 ")
    factorial = 1 
    for i in range (1,num+1):
        factorial = factorial*i
    return(factorial)

if __name__ =="__main__":
    try:
        user_input = int (input("Enter the value how you want factorial"))
        result = findfactoral(user_input)
        print(result)
    except (TypeError, ValueError) as e : 
        print(e) 
    except Exception as e : 
        print(e)


        
```
